```python
def pSpectraCharacterize(pSpectrDic,index,plot=False,color='b'):
    '''computes power spectrum characteristics 
    for a dictionary/array of power spectra'''
    
    
    slope1 = []
    slope2 = []
    S = []
    
    for i in index:
        try:
            dic = pinkNoiseCharacterize(pSpectrDic['spectra'][i])
        except:
            print "error index %s"%i
            continue
        
        slope1 = np.append(slope1,dic['slope1'])
        slope2 = np.append(slope2,dic['slope2'])
        S = np.append(S,pSpectrDic['S'][i])
              
    if plot:
        pl.subplot(131)
        fit = stats.linregress(slope1,S)
        print fit
        pl.plot(slope1,S,'.',c=color)
        pl.plot(slope1,slope1*fit[0]+fit[1],ms='-',c=color)
        pl.xlabel("slope1")
        pl.ylabel("normalized entropy")
        
        pl.subplot(132)
        pl.plot(slope2,S,'.',c=color)
        pl.xlabel("slope2")
        pl.ylabel("normalized entropy")
        
        pl.subplot(133)
        pl.plot(slope1,slope2,'.',c=color)
        pl.xlabel("slope1")
        pl.ylabel("slope2")
        
    return {'S':S,'slope1':slope1,'slope2':slope2}
```

```python
def pinkNoiseCharacterize(pspectrum,normalize=True,plot=False):
    '''Compute main power spectrum characteristics'''
    if normalize:
        pspectrum = pspectrum/np.sum(pspectrum)
    
    S = entropy(pspectrum,1)
    
    x = np.arange(1,len(pspectrum)+1)
    lx = np.log10(x)
    ly = np.log10(pspectrum)
    
    c1 = (x > 0)*(x < 80)
    c2 = x >= 80
    
    fit1 = stats.linregress(lx[c1],ly[c1])
    fit2 = stats.linregress(lx[c2],ly[c2])
    
    #print fit1
    #print fit2
    
    if plot:
        plot(lx,ly)
        plot(lx[c1],lx[c1]*fit1[0]+fit1[1],'r-')
        plot(lx[c2],lx[c2]*fit2[0]+fit2[1],'r-')
        
    return {'S':S,'slope1':fit1[0],'slope2':fit2[0]}
    
```

### Power Spectrum and Generalized Entropy

```python
def pSpectrum(vector):
    
    A = np.fft.fft(vector)
    ps = np.abs(A)**2
    ps = ps[:len(ps)/2]
        
    return ps

#ps = pSpectrum(test[500:600])
#loglog(ps)

def entropy(power_spectrum,q):
    q = float(q)
    
    power_spectrum = np.array(power_spectrum)
        
    if not q ==1:
        S = 1/(q-1)*(1-np.sum(power_spectrum**q))
    else:
        S = - np.sum(power_spectrum*np.log2(power_spectrum))
        
    return S

def entropySeries(time,rawData,q,windowSize=1,normalize=True):
    
    
    #t = tSeries['rawTime']
    #raw = tSeries['rawValue']
    
    t = time
    raw = rawData
    
    intervales = np.arange(np.trunc(min(t)+1),np.trunc(max(t)+1),windowSize)

    T = []
    S = [] 
    spectra = []
    
    for i,ix in enumerate(intervales[:-1]):
        #print i,ix
        
        c = (t>=ix)*(t<intervales[i+1])
        
        vector = raw[c]
        ps = np.array(pSpectrum(vector)) # Power Spectrum
        
        try:
            spectra += [ps]
        except:
            spectra = [ps]

        
        
        #if normalize:
        ps = ps/np.sum(ps)
        
        s = entropy(ps,q) #Compute Entropy
        S.append(s)
        T.append(i)

    S = np.array(S)
    T = np.array(T)
    
    if normalize:
        S = (S-np.mean(S))/np.std(S)
        
    return {'time' : list(intervales[:-1]), 'S': list(S),'spectra':spectra}


def avgPowerSpectrum(pspectra,xOutput):
    '''Computes an average power spectrum
    from multiple power spectra, and returns 
    (interpolated) evaluations for xOutput values '''
    
    if isinstance(pspectra,dict):
        pspectra = pspectra.values()
    
    l = len(pspectra)
    array = np.zeros([l,len(xOutput)])
    S = []
    
    for i,ps in enumerate(pspectra):
        #print i
        s = entropy(ps/np.sum(ps),1)
        S.append(s)
        
        x = np.arange(1,len(ps)+1)
        f = interp1d(x,ps/np.sum(ps))
        try:
            array[i] = f(xOutput)
        except:
            continue
            
    return array,S
```

### Synchronization Functions

```python
def syncTime(data):            
    subjects = data['subjects']
    
    minT = []
    maxT = []
    
    for s in subjects.keys():
        minT.append(np.min(subjects[s]['raw']['rawTime']))
        maxT.append(np.max(subjects[s]['raw']['rawTime']))
        
        #print subject,len(data[subject]['rawTime'])
        
    min_t = np.max(minT)
    max_t = np.min(maxT)
    
    print min_t,max_t
    
    for subject in subjects.keys():
        c = (np.array(subjects[subject]['raw']['rawTime']) > min_t)*(np.array(subjects[subject]['raw']['rawTime']) < max_t)
        subjects[subject]['raw']['rawTime'] = list(np.array(subjects[subject]['raw']['rawTime'])[c])
        subjects[subject]['raw']['rawValue'] = list(np.array(subjects[subject]['raw']['rawValue'])[c])

        #print subject,"%.2f , %.2f , %.2f"%(min(data[subject]['raw']['rawTime']),max(data[subject]['rawTime']),len(data[subject]['rawTime']))
    

    #data['subjects'] = subjects

    return data,min_t,max_t

def syncVoiceTime(date,session,t_min=0,t_max=0):
    #Load Data
    key = bucket.get_key("%s/%s/voice_manual_coding.csv"%(date,session))
    dataV = key.get_contents_as_string()
    dataV = re.sub(":",",",dataV)
    dataV = csv.reader(dataV.split("\r"),delimiter=",")
 
    #Process CSV file
    tOn = []
    tOff = []
    nameV = []

    for l,line in enumerate(dataV):
        
        if l==0:
            start = int(line[0])
        else:
            line[0:4]=map(int,line[0:4])
            
            t0 = start + int(line[0]*60+line[1])
            t1 = start + int(line[2]*60+line[3])
            name = line[4]
            
            tOn = np.append(tOn,t0)
            tOff = np.append(tOff,t1)
            nameV = np.append(nameV,name)
    
    if t_min == 0 and t_max ==0:
        t_min = np.min(tOn)
        t_max = np.max(tOff)

    t = np.arange(np.trunc(t_min)+1,np.trunc(t_max)+1,1)

    VoiceJson = {}

    for name in np.unique(nameV):
        #print name
        
        active = np.zeros_like(t)
        
        index = np.argwhere(name==nameV)   
        #print tOn[index],tOff[index]
        
        for i,to in enumerate(tOn[index]):
            #print name,int(to),int(tOff[index][i])
            iOn = np.argwhere((t>=to)*(t<tOff[index][i])) 
            active[iOn]=1
            
        VoiceJson[name] = {'time': list(t[:-1]), 'active' : list(active[:-1])}
    
    return VoiceJson


def syncEntropy(jsonData,q,normalize=True):
    
    subjects = jsonData['subjects']
    
    data = {}
    for subject in subjects.keys():
        subjects[subject] = makeArrays(subjects[subject])
        #print data[subject]['rawTime']#,data[subject]['rawValue']
        S = entropySeries(subjects[subject]['raw']['rawTime'],subjects[subject]['raw']['rawValue'],q,normalize=normalize)
        data[subject] = S
        subjects[subject] = makeLists(subjects[subject])
        
    jsonData['entropy'] = {str(q) : data}

    return jsonData
```

```python
''' S3 Handling Functions'''


def indexFiles():
    
    keys =bucket.list()

    dic = {}
    
    for k in fileCSV(keys):
        #print k.name
        date,session,subject = re.findall("(.*?)/(.*?)/(.*?)_\d\d*?",k.name)[0]
        
        try:
            dic[date][session][subject] = k.name  
        except:
            try:
                dic[date][session] = {subject: k.name}
            except:
                dic[date] = {session : {subject : k.name}}
                
    return dic

def fileKeys(keys):
    return (key for key in keys if not key.name.endswith('/'))

def fileCSV(keys):
    return (key for key in keys if key.name.endswith('.txt'))
```

```python
def processRawSignal(csvType):
    data = {'meditation' : {'mTime' : [], 'mValue' : []}, 'attention' : {'aTime' : [], 'aValue' : []}, 'raw' : {'rawTime' : [], 'rawValue' : []}}
    types = ['Meditation','Attention','Raw'] 

    for l, line in enumerate(csvType):
        #print l,line
        try:
            typ = line[-2][:-1]
        except:
            continue
            
        if typ=='Meditation':
            data['meditation']['mTime'].append(line[0][:-1])
            data['meditation']['mValue'].append(line[-1])
        
        if typ=="Attention":
            data['attention']['aTime'].append(line[0][:-1])
            data['attention']['aValue'].append(line[-1])
            
        if typ=="Raw":
            data['raw']['rawTime'].append(line[0][:-1])
            data['raw']['rawValue'].append(int(line[-1]))
    return data


def makeLists(data):
    
    data['meditation']['mTime'] = map(float,data['meditation']['mTime'])
    data['meditation']['mValue'] = map(float,data['meditation']['mValue'])
        
    data['attention']['aTime'] = map(float,data['attention']['aTime'])
    data['attention']['aValue'] = map(float,data['attention']['aValue'])
            
    data['raw']['rawTime'] = map(float,data['raw']['rawTime'])
    data['raw']['rawValue'] = map(float,data['raw']['rawValue'])
    
    return data
    
def makeArrays(data):
    
    data['meditation']['mTime'] = np.array(map(float,data['meditation']['mTime']))
    data['meditation']['mValue'] = np.array(map(float,data['meditation']['mValue']))
        
    data['attention']['aTime'] = np.array(map(float,data['attention']['aTime']))
    data['attention']['aValue'] = np.array(map(float,data['attention']['aValue']))
            
    data['raw']['rawTime'] = np.array(map(float,data['raw']['rawTime']))
    data['raw']['rawValue'] = np.array(map(float,data['raw']['rawValue']))
    
    return data
      
    
def loadSessionFromS3(date,session,loadJson=True):
    
    #print "destination : ", "%s/%s/%s/%s.json"%(date,session,name,subject)
    
    if loadJson:
        try:
            key = bucket.get_key("%s/%s/data.json"%(date,session))
            jsonData = key.get_contents_as_string()
            jsonData = json.loads(jsonData)
            return jsonData
        except:
            pass
    
    print "%s/%s/data.json not found, processing raw data"%(date,session)
    
    dStructure = indexFiles()
    
    jsonData = {'subjects' : {},'voice' : {}}
    
    for subject in dStructure[date][session].keys():
        key = bucket.get_key(dStructure[date][session][subject])
        
        #d = key.get_contents_as_string()
        f = StringIO.StringIO(key.get_contents_as_string())
        csvType = csv.reader(f,delimiter=' ')
        
        data = processRawSignal(csvType)
        data = makeLists(data)
        
        jsonData['subjects'][subject] = data 
        print subject,data.keys()
        
    '''
    if date=="Feb06_2014" and session=="Session_1":
        print "blah"
        jsonData['subjects']['Vaidy']['raw']['rawTime'] = list(np.array(jsonData['subjects']['Vaidy']['raw']['rawTime'])+748)
        jsonData['subjects']['Vaidy']['meditation']['mTime'] = list(np.array(jsonData['subjects']['Vaidy']['meditation']['mTime'])+748)
        jsonData['subjects']['Vaidy']['attention']['aTime'] = list(np.array(jsonData['subjects']['Vaidy']['attention']['aTime'])+748)
    '''
    
    print jsonData['subjects'][subject]['raw']['rawTime'][:10]
    #synchronize time
    try:
        jsonData,min_t,max_t = syncTime(jsonData)
    except:
        print "error when syncing Time"
        return jsonData

    print jsonData['subjects'][subject]['raw']['rawTime'][:10]    
    #load voice encoding and synchronize with brainwaves
    try:
        voiceJson = syncVoiceTime(date,session,t_min=min_t,t_max=max_t)
        jsonData['voice']=voiceJson
    except:
        print "error when syncing Voice"
        return jsonData
    
    print jsonData['subjects'][subject]['raw']['rawTime'][:10]
    #encode Entropy (Shannon Entropy)
    try:
        jsonData = syncEntropy(jsonData,1,normalize=True)
    except:
        print "error when syncing Entropy"
        return jsonData    
    
    print jsonData['subjects'][subject]['raw']['rawTime'][:10]
    #encode json and upload to S3
    try:
        jsonDataDump = json.dumps(jsonData)
        key = bucket.new_key("%s/%s/data.json"%(date,session))
        key.set_contents_from_string(jsonDataDump)
    except:
        print "error when encoding & uploading json to S3"
        return jsonData
    
    return jsonData
```
